```python
train=df[df['Date']<'201501010']
test=df[df['Date']>'20141231']
```

```python
train.head()
```

```python
test.head()
```

```python
#removing everything apart from a-z and A-Z.
data=train.iloc[:,2:27]
data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
data
```

```python
list1=[i for i in range(25)] #chaing the cloumns from strings to numbers.
new_index=[str(i) for i in list1]
data.columns=new_index
data.head()
```

```python
for i in new_index:
    data[i]=data[i].str.lower() #converting everything into lowercase.
data.head(2)
```

```python
#ive to make the all headlines in one row as one paragraph.
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tf_idf= TfidfVectorizer(stop_words='english', max_features=5000)
```

```python
#implement tf-idf
traindata=tf_idf.fit_transform(headlines)
```

```python
##doing modifications for test dataset.
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(y) for y in test.iloc[row,2:27]))
```

```python
test.head()
```

```python
testdata=tf_idf.transform(test_transform)
```

```python
#creating the model.
model=RandomForestClassifier(n_estimators=300,criterion='entropy')
model.fit(traindata,train['Label'])
```

```python
predicts=model.predict(testdata)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predicts)
print(matrix)
```

```python
score=accuracy_score(test['Label'],predicts)
print(score)
```
